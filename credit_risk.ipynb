```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['loan_int_rate'].fillna(df['loan_int_rate'].mean(),inplace = True)
```

```python
df['person_emp_length'].fillna(df['person_emp_length'].mean(),inplace = True)
```

```python
df.isnull().sum()
```

```python
y = df['loan_status']
x = df.drop(['loan_status'],axis = 1)
x = x.select_dtypes(include = 'number')
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_ts,y_tr,y_ts = train_test_split(x,y,random_state = 42 , test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(x_tr,y_tr)
pred_tr = model.predict(x_tr)
pred_ts = model.predict(x_ts)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_tr,pred_tr))
print(classification_report(y_ts,pred_ts))
```
